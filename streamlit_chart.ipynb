```python
df = pd.read_csv('./naver_place/keyword_rank_top5test1.csv')
```

```python
df.head(2)
df.rename(columns={'search': 'keyword'}, inplace=True)
```

```python
def modify_values(row):
    if row['hour'] == 4:
        return 13
    elif row['hour'] == 10:
        return 19
    else:
        return row['hour']

def modify_values2(row):
    if row['ko_hour'] == 13:
        return '오전'
    elif row['ko_hour'] == 19:
        return '오후'
    else:
        return row['ko_hour']
```

```python
dd = df.copy()
dd['ko_hour'] = dd.hour
dd['apm'] = dd.ko_hour

dd['ko_hour'] = dd.apply(modify_values, axis=1)

dd['apm'] = dd.apply(modify_values2, axis=1)


dd[['title','ko_hour','apm']]
```
